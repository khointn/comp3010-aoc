```python
level = 0
for char in input():
    if char == "(":
        level+=1
    elif char == ")":
        level-=1
print(level)
```

```python
level = 0
user = input()
for char in range(len(user)):
    if user[char] == "(":
        level+=1
    elif user[char] == ")":
        level-=1
    if(level == -1):
      print(char+1)
      break
```
